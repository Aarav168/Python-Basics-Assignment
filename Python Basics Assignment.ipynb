```python
# 2.Describe the role of predefined keywords in Python and provide examples of how they are used in a program?


def keyword_explanation():
    explanation = (
        "In Python, predefined keywords (also known as reserved words) are words that have special meaning. "
        "They are part of the syntax and cannot be used as identifiers (like variable or function names). "
        "These keywords control the structure and flow of a program."
    )
    print(explanation)
    
    # Example usage of some common keywords
    print("\nExamples of Python keywords:")

    # if-else keyword usage
    x = 10
    if x > 5:
        print("x is greater than 5")  # using 'if'
    else:
        print("x is less than or equal to 5")  # using 'else'

    # for keyword usage
    for i in range(3):  # using 'for' and 'in'
        print(f"Loop iteration {i}")

    # def and return keyword usage
    def greet(name):  # using 'def'
        return f"Hello, {name}!"  # using 'return'

    print(greet("Alice"))

    # import keyword usage
    import math  # using 'import'
    print("Square root of 16 is:", math.sqrt(16))

# Call the function
keyword_explanation()
```

```python
# 3.Compare and contrast mutable and immutable objects in Python with examples?

def mutable_vs_immutable():
    print("üîÅ Mutable Objects:")
    print("- Can be changed after creation.")
    print("- Examples: list, dict, set")

    # Mutable Example: List
    my_list = [1, 2, 3]
    print("Original List:", my_list)
    my_list.append(4)
    print("Modified List (after append):", my_list)

    print("\nüîí Immutable Objects:")
    print("- Cannot be changed after creation.")
    print("- Examples: int, float, string, tuple")

    # Immutable Example: String
    my_string = "hello"
    print("Original String:", my_string)
    new_string = my_string.replace("h", "y")
    print("Modified String (using replace):", new_string)
    print("Original String remains unchanged:", my_string)

    # Immutable Example: Tuple
    my_tuple = (1, 2, 3)
    print("Tuple:", my_tuple)
    try:
        my_tuple[0] = 100  # This will raise an error
    except TypeError as e:
        print("Error when trying to modify tuple:", e)

# Call the function
mutable_vs_immutable()
```

```python
# 4.Discuss the different types of operators in Python and provide examples of how they are used?

def python_operators():
    print("üîπ 1. Arithmetic Operators")
    a, b = 10, 3
    print("Addition:", a + b)        # +
    print("Subtraction:", a - b)     # -
    print("Multiplication:", a * b)  # *
    print("Division:", a / b)        # /
    print("Modulus:", a % b)         # %
    print("Exponent:", a ** b)       # **
    print("Floor Division:", a // b) # //

    print("\nüîπ 2. Comparison Operators")
    print("Equal:", a == b)          # ==
    print("Not Equal:", a != b)      # !=
    print("Greater Than:", a > b)    # >
    print("Less Than:", a < b)       # <
    print("Greater or Equal:", a >= b)  # >=
    print("Less or Equal:", a <= b)     # <=

    print("\nüîπ 3. Logical Operators")
    x, y = True, False
    print("AND:", x and y)           # and
    print("OR:", x or y)             # or
    print("NOT:", not x)             # not

    print("\nüîπ 4. Assignment Operators")
    num = 5
    num += 2  # same as num = num + 2
    print("+= Operator:", num)
    num *= 3  # same as num = num * 3
    print("*= Operator:", num)

    print("\nüîπ 5. Bitwise Operators")
    x, y = 5, 3  # binary: 101 & 011
    print("AND (&):", x & y)
    print("OR (|):", x | y)
    print("XOR (^):", x ^ y)
    print("Left Shift (<<):", x << 1)
    print("Right Shift (>>):", x >> 1)

    print("\nüîπ 6. Membership Operators")
    lst = [1, 2, 3, 4]
    print("3 in list:", 3 in lst)
    print("5 not in list:", 5 not in lst)

    print("\nüîπ 7. Identity Operators")
    a = [1, 2]
    b = a
    c = [1, 2]
    print("a is b:", a is b)         # True (same object)
    print("a is c:", a is c)         # False (same value, different objects)
    print("a == c:", a == c)         # True (same content)

# Call the function
python_operators()
```

```python
# 5.Explain the concept of type casting in Python with examples?

def type_casting_examples():
    print("üìò Type Casting: Converting one data type to another.")

    print("\nüîπ 1. Implicit Type Casting (Python does it automatically)")
    int_num = 10
    float_num = 2.5
    result = int_num + float_num  # int is automatically converted to float
    print("int_num + float_num =", result)
    print("Type of result:", type(result))

    print("\nüîπ 2. Explicit Type Casting (Done manually by the programmer)")

    # int() ‚Äì Converts to integer
    x = "100"
    int_x = int(x)
    print("String to int:", int_x, "| Type:", type(int_x))

    # float() ‚Äì Converts to float
    y = "12.34"
    float_y = float(y)
    print("String to float:", float_y, "| Type:", type(float_y))

    # str() ‚Äì Converts to string
    z = 456
    str_z = str(z)
    print("Integer to string:", str_z, "| Type:", type(str_z))

    # list(), tuple(), set() ‚Äì Convert between collections
    my_tuple = (1, 2, 3)
    my_list = list(my_tuple)
    print("Tuple to List:", my_list)

    my_list2 = [4, 5, 6]
    my_set = set(my_list2)
    print("List to Set:", my_set)

# Call the function
type_casting_examples()
```

```python
# 6.How do conditional statements work in Python? Illustrate with examples?

def conditional_statements():
    print("üìò Conditional statements are used to make decisions in Python.")

    print("\nüîπ if statement")
    num = 10
    if num > 0:
        print("num is positive")

    print("\nüîπ if-else statement")
    num = -5
    if num >= 0:
        print("num is non-negative")
    else:
        print("num is negative")

    print("\nüîπ if-elif-else statement")
    score = 85
    if score >= 90:
        print("Grade: A")
    elif score >= 80:
        print("Grade: B")
    elif score >= 70:
        print("Grade: C")
    else:
        print("Grade: F")

    print("\nüîπ Nested if statement")
    age = 20
    if age >= 18:
        if age < 21:
            print("You are an adult but not allowed to drink in the US.")
        else:
            print("You are an adult and allowed to drink.")
    else:
        print("You are underage.")

# Call the function
conditional_statements()
```

```python
# 7.Describe the different types of loops in Python and their use cases with examples?

def python_loops():
    print("üîÅ Loops in Python are used to execute a block of code repeatedly.")

    print("\nüîπ 1. for loop")
    print("Use Case: Iterating over a sequence like list, tuple, string.")
    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print("Fruit:", fruit)

    print("\nüîπ 2. for loop with range()")
    print("Use Case: Repeating code a specific number of times.")
    for i in range(3):
        print("Loop iteration:", i)

    print("\nüîπ 3. while loop")
    print("Use Case: Run a block of code as long as a condition is true.")
    count = 0
    while count < 3:
        print("Count is:", count)
        count += 1

    print("\nüîπ 4. Nested Loops")
    print("Use Case: Loop inside another loop (e.g., 2D structures).")
    for i in range(2):
        for j in range(3):
            print(f"i={i}, j={j}")

    print("\nüîπ 5. Loop Control Statements")
    print("Use Case: Modify loop behavior using break, continue, pass")

    print("\nüî∏ break example:")
    for i in range(5):
        if i == 3:
            break
        print(i)

    print("\nüî∏ continue example:")
    for i in range(5):
        if i == 2:
            continue
        print(i)

    print("\nüî∏ pass example:")
    for i in range(3):
        pass  # Placeholder when no action is needed
    print("Loop with 'pass' ran successfully.")

# Call the function
python_loops()
```
